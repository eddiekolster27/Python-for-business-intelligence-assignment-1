## Task 1: Transaction Filtering & Regional Revenue Analysis

```python
transactions = [
    {"id": 1, "region": "North", "price": 120.0, "quantity": 3},
    {"id": 2, "region": "South", "price": 0, "quantity": 5},
    {"id": 3, "region": "East",  "price": 85.50, "quantity": -2},
    {"id": 4, "region": "North", "price": 200.0, "quantity": 1},
    {"id": 5, "region": "South", "price": 50.0, "quantity": 10},
    {"id": 6, "region": "East",  "price": 310.0, "quantity": 2},
    {"id": 7, "region": "West",  "price": 175.0, "quantity": 4},
    {"id": 8, "region": "West",  "price": 90.0, "quantity": 6},
    {"id": 9, "region": "North", "price": 60.0, "quantity": 8},
    {"id": 10, "region": "South", "price": 45.0, "quantity": 0},
]

# Filter out invalid transactions (price == 0 or quantity < 0)
valid_transactions = [
    t for t in transactions
    if t["price"] != 0 and t["quantity"] >= 0
]

print(f"Valid transactions: {len(valid_transactions)} out of {len(transactions)}")

for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# Group revenue by region
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]

print("\nRevenue by region:")
for region, revenue in sorted(revenue_by_region.items()):
    print(f"  {region}: {revenue:,.2f}")

highest_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"\nHighest revenue region: {highest_region} ({revenue_by_region[highest_region]:,.2f})")
```

## Task 2: Sales Summary Function

```python
def summarise_sales(daily_sales):
    """Return a summary dict for a list of daily sales figures."""
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)

    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1

    above_average_count = sum(1 for s in daily_sales if s > average)

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count,
    }


daily_sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

summary = summarise_sales(daily_sales)

print("Sales Summary")
for key, value in summary.items():
    print(f"  {key}: {value}")
```

## Task 3: Student Records — Grading, Attendance & Highest Grade

```python
students = [
    {"name": "Alice",   "grade": 92, "attendance": 0.95},
    {"name": "Bob",     "grade": 47, "attendance": 0.60},
    {"name": "Charlie", "grade": 73, "attendance": 0.88},
    {"name": "Diana",   "grade": 85, "attendance": 0.78},
    {"name": "Eve",     "grade": 38, "attendance": 0.55},
    {"name": "Frank",   "grade": 65, "attendance": 0.92},
    {"name": "Grace",   "grade": 51, "attendance": 0.81},
]

passing_students = []

for s in students:
    name, grade = s["name"], s["grade"]

    if grade >= 80:
        status = "Excellent"
    elif grade >= 50:
        status = "Pass"
    else:
        status = "Fail"

    print(f"{name}: {grade} — {status}")

    if grade >= 50:
        passing_students.append(s)

print(f"\nPassing students: {[s['name'] for s in passing_students]}")

low_attendance_count = sum(1 for s in students if s["attendance"] < 0.80)
print(f"Students with attendance < 80%: {low_attendance_count}")

# Find highest grade with a manual loop
highest_grade_loop = students[0]["grade"]
highest_name_loop = students[0]["name"]
for s in students[1:]:
    if s["grade"] > highest_grade_loop:
        highest_grade_loop = s["grade"]
        highest_name_loop = s["name"]

print(f"\nHighest grade (loop):  {highest_name_loop} with {highest_grade_loop}")

# Find highest grade using max()
top_student = max(students, key=lambda s: s["grade"])
print(f"Highest grade (max()): {top_student['name']} with {top_student['grade']}")
```

## Task 4: Portfolio Class

```python
class Portfolio:
    """Simple investment portfolio tracker."""

    def __init__(self):
        self.assets = []

    def add_asset(self, name, value):
        self.assets.append({"name": name, "value": value})

    def total_value(self):
        return sum(a["value"] for a in self.assets)

    def most_valuable(self):
        return max(self.assets, key=lambda a: a["value"])

    def __str__(self):
        lines = ["Portfolio:"]
        for a in self.assets:
            lines.append(f"  {a['name']}: {a['value']:,.2f}")
        lines.append(f"  Total: {self.total_value():,.2f}")
        return "\n".join(lines)


portfolio = Portfolio()
portfolio.add_asset("AAPL", 15_200.00)
portfolio.add_asset("MSFT", 22_750.50)
portfolio.add_asset("TSLA", 8_430.75)
portfolio.add_asset("AMZN", 18_900.00)

print(portfolio)

top_asset = portfolio.most_valuable()
print(f"\nMost valuable: {top_asset['name']} ({top_asset['value']:,.2f})")
```

## Task 5: Order Date Parsing & Date Range

```python
from datetime import datetime

order_date_strings = [
    "2024-01-15",
    "2024-03-22",
    "2024-02-08",
    "2024-05-30",
    "2024-04-12",
]

# Parse strings into datetime objects
order_dates = [datetime.strptime(d, "%Y-%m-%d") for d in order_date_strings]

earliest_order = min(order_dates)
latest_order = max(order_dates)
days_between = (latest_order - earliest_order).days

print(f"Earliest order: {earliest_order.date()}")
print(f"Latest order:   {latest_order.date()}")
print(f"Days between:   {days_between}")
```

## Task 6: NumPy — Quarterly Revenue Analysis

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 260],   # 2022
    [240, 250, 220, 280],   # 2023
    [270, 290, 250, 310],   # 2024
])

print("Quarterly revenue (millions):")
print(revenue)

annual_totals = revenue.sum(axis=1)
print(f"\nAnnual totals: {annual_totals}")

quarterly_means = revenue.mean(axis=0)
print(f"Quarterly means: {quarterly_means}")

# Apply 5% growth adjustment to the 2024 row
revenue_adjusted = revenue.copy().astype(float)
revenue_adjusted[2] *= 1.05
print(f"\n2024 row after 5% growth adjustment: {revenue_adjusted[2]}")

# Values greater than 250
high_revenue = revenue[revenue > 250]
print(f"\nValues > 250: {high_revenue}")
```

## Task 7: Pandas DataFrame — Employee Analysis

```python
import pandas as pd

employees = pd.DataFrame({
    "name":        ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "department":  ["Sales", "Engineering", "Sales", "HR", "Engineering"],
    "salary":      [55000, 78000, 62000, 48000, 91000],
    "years_exp":   [3, 7, 5, 2, 10],
    "performance": [4.2, 3.8, 4.5, 3.1, 4.9],
})

print("Original DataFrame:")
print(employees)

# Add salary per year of experience column
employees["salary_per_year_exp"] = (employees["salary"] / employees["years_exp"]).round(2)
print("\nWith salary_per_year_exp:")
print(employees)

# Sort by performance descending
sorted_by_perf = employees.sort_values("performance", ascending=False)
print("\nSorted by performance (desc):")
print(sorted_by_perf)

# .loc filter: Engineering employees with salary > 70000
eng_high_salary = employees.loc[
    (employees["department"] == "Engineering") & (employees["salary"] > 70000)
]
print("\nEngineering employees with salary > 70,000:")
print(eng_high_salary)

# .rename columns
renamed = employees.rename(columns={"name": "employee_name", "years_exp": "experience"})
print("\nRenamed columns:")
print(renamed.columns.tolist())

# .iloc selection: first 3 rows, columns 0–2
subset = employees.iloc[:3, :3]
print("\n.iloc[:3, :3]:")
print(subset)
```

## Task 8: Random Data Generation & Computed Columns

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)

n_days = 30

# Generate date index
date_index = pd.date_range(start="2024-01-01", periods=n_days, freq="D")

# Generate random data
visitors = rng.integers(500, 3000, size=n_days)
conversion_rate = rng.uniform(0.01, 0.08, size=n_days).round(4)
avg_order_value = rng.uniform(20, 120, size=n_days).round(2)

df = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}, index=date_index)

# Compute conversions and daily revenue
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round(0).astype(int)
df["daily_revenue"] = (df["conversions"] * df["avg_order_value"]).round(2)

print("E-commerce daily metrics (first 10 rows):")
print(df.head(10))

print(f"\nTotal revenue over {n_days} days: {df['daily_revenue'].sum():,.2f}")
print(f"Best day: {df['daily_revenue'].idxmax().date()} ({df['daily_revenue'].max():,.2f})")
print(f"Average daily revenue: {df['daily_revenue'].mean():,.2f}")
```
